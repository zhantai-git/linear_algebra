```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Algebraic and geometric interpretations

```python
# 2-dimensional vector
v2 = [3, -2]

# 3-dimensional vector
v3 = [4, -3, 2]

# row to column (or vice-verca):
v3t = np.transpose(v3)


# plot them
v2 = [1, 2]
plt.plot([0,v2[0]],[0,v2[1]])

plt.axis('equal')
plt.plot([-4, 4],[0, 0], 'k--')
plt.plot([0, 0],[-4, 4], 'k--')
plt.grid()
plt.axis((-4, 4, -4, 4))
plt.show()
```

```python
# plot in 3d
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')
ax.plot([0, v3[0]],[0, v3[1]],[0, v3[2]], linewidth=3)

# make the plot look nicer
ax.plot([0, 0],[0, 0],[-4, 4], 'k--')
ax.plot([0, 0],[-4, 4], [0, 0], 'k--')
ax.plot([-4, 4],[0, 0], [0, 0], 'k--')
plt.show()
```

### Vector addition and substraction

```python
# two vectors in R2
v1 = np.array([3, -1])
v2 = np.array([2, 4])

v3 = v1 + v2

# plot them
plt.plot([0,v1[0]], [0,v1[1]],'b',label='v1')
plt.plot([0,v2[0]]+v1[0], [0,v2[1]]+v1[1],'r',label='v2')
plt.plot([0,v3[0]], [0,v3[1]],'k',label='v1+v2')

plt.legend()
plt.axis('square')
plt.axis((-6, 6, -6, 6))
plt.grid()
plt.show()
```

```python
v_1 = np.array([1, 1, 2])
v_2 = np.array([1, 2, 4])
v_3 = v_1 + v_2


# plot in 3d
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')
ax.plot([0, v_3[0]],[0, v_3[1]],[0, v_3[2]], linewidth=3)
ax.plot([0, v_1[0]],[0, v_1[1]],[0, v_1[2]], linewidth=3)
ax.plot([v_1[0], v_2[0]+v_1[0]],[v_1[1], v_2[1]+v_1[1]],[v_1[2], v_2[2]+v_1[2]], linewidth=3)
# make the plot look nicer
ax.plot([0, 0],[0, 0],[-6, 6], 'k--')
ax.plot([0, 0],[-6, 6], [0, 0], 'k--')
ax.plot([-6, 6],[0, 0], [0, 0], 'k--')
plt.show()
```

```python
import plotly.graph_objects as go

# 3d with plotly
v_1 = np.array([1, 1, 2])
v_2 = np.array([1, 2, 4])
v_3 = v_1 + v_2
#    traces
fig = go.Figure()

#  v_1
fig.add_trace(go.Scatter3d(
    x=[0, v_1[0]], 
    y=[0, v_1[1]], 
    z=[0, v_1[2]],
    mode='lines',
    line=dict(color='red', width=5),
    name='v_1'
))

#  v_2   v_1
fig.add_trace(go.Scatter3d(
    x=[v_1[0], v_2[0]+v_1[0]], 
    y=[v_1[1], v_2[1]+v_1[1]], 
    z=[v_1[2], v_2[2]+v_1[2]],
    mode='lines',
    line=dict(color='green', width=5),
    name='v_2'
))
#  v_3 
fig.add_trace(go.Scatter3d(
    x=[0, v_3[0]], 
    y=[0, v_3[1]], 
    z=[0, v_3[2]],
    mode='lines',
    line=dict(color='blue', width=5),
    name='v_3'
))

#  
fig.add_trace(go.Scatter3d(
    x=[-6, 6], y=[0, 0], z=[0, 0],
    mode='lines',
    line=dict(color='black', width=2, dash='dash'),
    showlegend=False
))

fig.add_trace(go.Scatter3d(
    x=[0, 0], y=[-6, 6], z=[0, 0],
    mode='lines',
    line=dict(color='black', width=2, dash='dash'),
    showlegend=False
))

fig.add_trace(go.Scatter3d(
    x=[0, 0], y=[0, 0], z=[-6, 6],
    mode='lines',
    line=dict(color='black', width=2, dash='dash'),
    showlegend=False
))

#   
fig.update_layout(
    scene=dict(
        xaxis=dict(range=[-6, 6]),
        yaxis=dict(range=[-6, 6]),
        zaxis=dict(range=[-6, 6]),
        aspectmode='cube'
    ),
    width=800,
    height=800
)

fig.show()
```
